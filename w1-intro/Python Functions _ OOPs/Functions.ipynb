## Problem if we don't use function

```python
#Initializing Variables
n=5
r=3

#Calculating factorial for n
n_fact=1
for i in range(1,n+1):
    n_fact*=i

#Calculating factorial for r
r_fact=1
for i in range(1,r+1):
    r_fact*=i

#Calculating factorial for n-r
n_r_fact=1
for i in range(1,n-r+1):
    n_r_fact*=i

#Calculating combination
nCr=n_fact//((n_r_fact)*r_fact)

nCr
```

## Using Function

```python
def fact(n): # the parameters the function will use are called, formal parameteres
    """ Function for calculating factorial"""
    factorial=1
    for i in range(1,n+1):
        factorial*=i
    return factorial
```

```python
#print - In-built Function to print output to the console
print(fact(5)) # the parameters we are passing while calling, it's called actual parameters
```

```python
#Calculating combination using factorial function
n=5
r=3
n_fact=fact(n)
r_fact=fact(r)
n_r_fact=fact(n-r)
nCr=n_fact//(n_r_fact*r_fact)
nCr
```

```python
def nCr(n,r):
    """ Function for calculating combination"""
    return fact(n)//(fact(n-r)*fact(r))
```

```python
#Function Calling
print(nCr(5,3))
```

```python
print(nCr(5,4))
```

```python
#To get info about the function
print(nCr.__doc__)
```

##Function With Recursive Paradigm

```python
def fact(n):
    if n == 0: #Base Condition
        return 1
    return n * fact(n-1)

fact(5)
```

## Parameters

###Parameters are of two types:
* Formal Parameters - Parameters which are being specified during the definition of the function.
* Actal Parameters - Parameters which are being specified during the function call.

##Actual Parameters are actually of four types :

* Positional Parameters
* Keyword Parameters
* Default Parameters
* Variable length Parameters



```python
#Function for adding two number
def add(a,b):
    return a+b
```

```python
#Positional Argument
print(add(2,7))
```

```python
print(add(2))
```

```python
#Default Argument
def add(a,b=10):
    return a+b
```

```python
print(add(1,2))
```

```python
print(add(1))
```

```python
#Keyword Argument
print(add(b=5,a=2))
```

```python
#1 Positional Argument and 1 Keyword Argument
print(add(10,b=5))
```

```python
#Variable Length Parameters
def print_name(*args):  #args will convert all input into a tuple
    for name in args:
        print(name)

print_name('Ram','Shyam')
```

```python
def print_student_details(**args):  #args will convert all input into a dictionary
    for key,value in args.items():
        print(key,value)

print_student_details(name='Ram',marks=100,grade='A')
```

```python
# linear search
def search(lst,x):
    """
    This function will return True if element is found otherwise return False
    lst- list
    x-integer
    """
    for i in lst:
        if i==x:
            return True
    return False
```

##Filter (Inbuilt-Function)
* It extracts elements from an iterable (list, tuple etc.) for which a function returns True.

```python
# Program to filter out only the even items from a list
new_list = list(filter(lambda x: (x%2 != 0) , lst))
print(new_list)
```

```python
def checkForOdd(num):
    return True if num%2 !=0 else False

new_list = list(filter(checkForOdd,lst))
print(new_list)
```

##Map (Inbuilt-Function)
* It applies a given function to each item of an iterable (list, tuple etc.) and returns an iterator.

```python
# Program to double each item in a list using map()
new_list = list(map(lambda x: x * 2 , lst))
print(new_list)
```

```python
def doubleNum(num):
    return num * 2
new_list = list(map(doubleNum,lst))
print(new_list)
```
