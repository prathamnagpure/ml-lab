# Reading from a file,Writing to a file.

Open a File on the Server
The open() function returns a file object, which has a read() method for reading the content of the file:

```python
#Specifying file path
base_file_path = "drive/MyDrive/Colab Notebooks/ML Lab (Aug'22=Dec'22)/Week-I/"

f = open(base_file_path + "demo.txt", "r")
print(f.read())
```

```python
# return the 5 fist characters of the file:
f = open(base_file_path + "demo.txt", "r")
print(f.read(5))
```

# Read Lines

```python
# You can return one line by using the readline() method:
f = open(base_file_path + "demo.txt", "r")
print(f.readline())
```

```python
# By calling readline() two times, you can read the two first lines:
# Read two lines of the file:
f = open(base_file_path + "demo.txt", "r")
print(f.readline())
print(f.readline())
```

###By looping through the lines of the file, you can read the whole file, line by line:

```python
f = open(base_file_path + "demo.txt", "r")
for line in f:
  print(line)
```

```python
# Close the file when you are finish with it:
f = open(base_file_path + "demo.txt", "r")
print(f.readline())
f.close()
```

# File Writing

To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content

```python
# Open the file "bootcamp.txt" and append content to the file:
demo_file_path = base_file_path + "demo.txt"
f = open(demo_file_path, "a")
f.write("Now the file has more content!")
f.close()
```

```python
#open and read the file after the appending:
f = open(demo_file_path, "r")
for line in f:
    print(line, end = '')
```

```python
# Open the file "bootcamp.txt" and overwrite the content:

f = open(demo_file_path, "w")
f.write("Woops! I have deleted the content!")
f.close()

#open and read the file after the appending:
f = open(demo_file_path, "r")
print(f.read())
```

Note: the "w" method will overwrite the entire file.
